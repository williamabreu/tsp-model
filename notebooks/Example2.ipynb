```python
from tspmodel.iohandler.FileInput import FileInput
from tspmodel.algorithm.TspSolverParticular import TspSolverParticular
```

```python
from tspmodel.graph.UndirectedGraph import UndirectedGraph
```

```python
from tspmodel.graph.Edge import Edge
```

```python
from tspmodel.util.Geometry import distance_vertex
```

```python
vertices = FileInput('tspmodel/tests/testcases/input4.txt').read()
```

```python
graph = UndirectedGraph()
for v in vertices:
    graph.add_vertex(v)

graph
```

```python
def run(vertices, path, graph):
    solution = 0.0
    for i in range(len(path) - 1):
        u = vertices[path[i]]
        v = vertices[path[i+1]]
        w = distance_vertex(u, v)
        graph.add_edge(Edge(u, v, w))
        solution += w
    return solution

path1 = (0, 1, 3, 4, 11, 12, 14, 15, 16, 18)
path2 = (18, 17, 13, 10, 9, 8, 7, 6, 5, 2, 0)

solution = run(vertices, path1, graph) + run(vertices, path2, graph)

graph
```

```python
display(HTML(f'<h3><i>Solution:</i> <code><strong>{solution:.2f}</strong></code></h3>'))
```
